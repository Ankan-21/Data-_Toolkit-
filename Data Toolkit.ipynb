3.Explain the following terms:
 The difference in np.array, np.asarray and np.asanyarray.
 The difference between Deep copy and shallow copy.


 Ans-

  numpy offers several functions for creating array-like objects, and it's essential to know the differences between np.array, np.asarray, and np.asanyarray. Here's a breakdown:

np.array
Purpose: This function creates a new NumPy array object.

Behavior: It always returns an ndarray, even if the input is already an ndarray. This means it will create a copy of the input data unless the copy parameter is set to False.

Use case: Use np.array when we want to ensure that we have a new array object, regardless of the input type.

import numpy as np  

x = np.array([1, 2, 3])  # Creates a new ndarray  
y = np.array(np.array([4, 5, 6]))  # Creates a new ndarray (copy)  
np.asarray
Purpose: This function converts the input into an array but does not create a copy if the input is already an ndarray.

Behavior: If the input is an ndarray, it will return it directly without copy. If it's not, it will create a new ndarray.

Use case: Use np.asarray when we want to avoid unnecessary copies of existing arrays.

x = np.asarray([1, 2, 3])  # Creates a new ndarray  
y = np.asarray(np.array([4, 5, 6]))  # Returns the existing ndarray, no new copy  
np.asanyarray
Purpose: This function converts the input to an array but will return subclasses of ndarray (such as maskedarray or matrix) if the input is of such types.
Behavior: Similar to np.asarray, but it does not restrict the output to be just an ndarray.
Use case: Use np.asanyarray if we want to preserve subclasses of ndarray.

import numpy as np  

class CustomArray(np.ndarray):  


x = np.asanyarray([1, 2, 3])  # Creates a new ndarray  
y = np.asanyarray(CustomArray([4, 5, 6]))  # Returns the existing CustomArray, no new copy  

Summary
np.array: Always creates a new ndarray.

np.asarray: Converts to ndarray but does not copy when the input is already an ndarray.

np.asanyarray: Similar to asarray, but it can return subclasses of ndarray.




     ##The difference between Deep copy and shallow copy.

The difference between a deep copy and a shallow copy primarily revolves around how they handle the duplication of objects and their nested structures in memory. Here’s a detailed explanation of both concepts:

Shallow Copy
Definition: A shallow copy creates a new object but does not create deep copies of nested objects (i.e., objects contained within the original object). Instead, it copies references to those nested objects.

Behavior: Changes made to the nested objects in the original will reflect in the shallow copy because both the original and the shallow copy reference the same nested objects.

Usage: Shallow copies are useful when we want to duplicate an object but do not need to create entirely independent nested objects.
Example in Python:

import copy  

original = [1, 2, [3, 4]]  
shallow_copied = copy.copy(original)  

# Modify the nested list in the original  
original[2][0] = 'Modified'  

print(original)         # Output: [1, 2, ['Modified', 4]]  
print(shallow_copied)  # Output: [1, 2, ['Modified', 4]]  
In this example, modifying the nested list in original will also reflect in shallow_copied, because both lists share the reference to the same nested list.


Deep Copy
Definition: A deep copy creates a new object and recursively copies all objects found in the original object, including all nested objects. As a result, the deep copy is entirely independent of the original.

Behavior: Changes made to the nested objects in the original do not affect the deep copy, as they contain their own copies of the nested objects.

Usage: Deep copies are useful when we need complete independence from the original object and its nested structures.
Example in Python:

import copy  

original = [1, 2, [3, 4]]  
deep_copied = copy.deepcopy(original)  

# Modify the nested list in the original  
original[2][0] = 'Modified'  

print(original)         # Output: [1, 2, ['Modified', 4]]  
print(deep_copied)     # Output: [1, 2, [3, 4]]  
In this case, modifying the nested list in original does not affect deep_copied, since they are completely separate objects in memory.

Summary
Shallow Copy:
Only the top-level object is copied.
Nested objects are referenced, not copied.
Changes to nested objects in the original affect the shallow copy.

Deep Copy:
The entire object structure, including nested objects, is recursively copied.
Changes to nested objects in the original do not affect the deep copy.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create two random arrays
np.random.seed(0)  # For reproducibility
x = np.random.randn(100)
y = np.random.randn(100)

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y})

# Determine the quadrant for each point
data['quadrant'] = np.where(data['x'] >= 0,
                            np.where(data['y'] >= 0, 'Quadrant I', 'Quadrant IV'),
                            np.where(data['y'] >= 0, 'Quadrant II', 'Quadrant III'))

# Create the scatter plot
plt.figure(figsize=(10, 6))
palette = {'Quadrant I': 'blue', 'Quadrant II': 'green', 'Quadrant III': 'red', 'Quadrant IV': 'orange'}
scatter = sns.scatterplot(data=data, x='x', y='y', hue='quadrant', palette=palette)

# Adding legend and labels
plt.legend(title='Quadrants')
plt.title('Quadrant-wise Scatter Plot', fontsize=16)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Generate data
data = np.random.normal(loc=0, scale=1, size=1000)  # Normal distribution with mean=0, std=1

# Step 2: Create the histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# Step 3: Overlay the PDF
xmin, xmax = plt.xlim()  # Get the current x-axis limits
x = np.linspace(xmin, xmax, 100)  # Generate values for x-axis
p = norm.pdf(x, loc=0, scale=1)   # Calculate the PDF

plt.plot(x, p, 'k', linewidth=2)  # Overlay the PDF

# Step 4: Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Step 5: Set the title
plt.title('Histogram with PDF Overlay')

# Step 6: Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a date range
date_range = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')

# Generate random temperature and humidity data
np.random.seed(0)  # For reproducibility
temperature = np.random.uniform(low=15, high=35, size=len(date_range))  # Temperature in °C
humidity = np.random.uniform(low=30, high=100, size=len(date_range))  # Humidity in %

# Create a DataFrame
data = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# Display the first few rows of the dataset
print(data.head())



# Create the plot
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot Temperature
ax1.scatter(data['Date'], data['Temperature'], color='red', label='Temperature', alpha=0.6)
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.scatter(data['Date'], data['Humidity'], color='blue', label='Humidity', alpha=0.6)
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Set the title
plt.title('Temperature and Humidity Over Time')

# Show the plot
plt.show()
```

```python
import numpy as np

# Step 1: Create a 3D NumPy array of shape (3, 3, 3) containing random integers between 1 and 10
array_3d = np.random.randint(1, 11, size=(3, 3, 3))

# Print the generated array
print("Generated 3D Array:")
print(array_3d)

# Step 2: Find the indices of the maximum values along each depth level (third axis)
max_indices = np.argmax(array_3d, axis=2)

# Print the max indices
print("\nIndices of Maximum Values along the Depth Level:")
print(max_indices)

# Step 3: Perform element-wise multiplication of the array with itself
elementwise_product = array_3d * array_3d

# Print the result of element-wise multiplication
print("\nElement-wise Multiplication Result:")
print(elementwise_product)
```

```python
import numpy as np

# Create a NumPy array with random integers between 1 and 10 of shape (5, 6)
array = np.random.randint(1, 11, size=(5, 6))
print("Original Array:")
print(array)

# a) Extract all even integers from array
even_integers = array[array % 2 == 0]
print("\nEven Integers:")
print(even_integers)

# b) Extract all odd integers from array
odd_integers = array[array % 2 != 0]
print("\nOdd Integers:")
print(odd_integers)
```
